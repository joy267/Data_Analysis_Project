```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\Users\\Mrityunjoy Mandal\\OneDrive\\Desktop\\supermarket_sales.csv")
```

```python
df.head(11)
```

# 1. which city has more female shoppers?

```python
df.groupby(['Gender','City']).count()
```

```python
female_shoppers = df.groupby(['Gender','City']).count()['Customer type']
```

```python
female_shoppers
```

```python
female_shoppers.unstack(level=0).plot(kind='bar')
plt.legend(loc='lower left')
```

# 2. Who spends more? Men or Female

```python
df.groupby(['Gender']).sum()['Total']
```

```python
total = df.groupby(['Gender']).sum()['Total']
total.plot(kind='bar')
```

# 3. Which type of customer spends more? Member or Normal

```python
df.groupby(['Customer type']).count()['Total']
```

```python
customer = df.groupby(['Customer type']).count()['Total']
customer.plot(kind='bar')
plt.legend()
```

# 4. Which product line has sells more?

```python
df.groupby(['Product line']).count()['Total']
```

```python
product = df.groupby(['Product line']).count()['Total']
product.plot(kind='bar')
plt.grid(axis='y')
```

# 5. Which Product line is popular among Men and Female?

```python
df.groupby(['Gender', 'Product line']).count()['Total']
```

```python
popular = df.groupby(['Gender', 'Product line']).count()['Total']
popular.unstack(level=1).plot(kind='bar')
plt.grid(axis='y')
```

# 6. Which days of the month has most sales?

```python
df['Date']
```

```python
pd.to_datetime(df['Date'])  # Converting data type object to datetime format
```

```python
pd.to_datetime(df['Date']).dt.day
pd.to_datetime(df['Date']).dt.month
df['Day'] = pd.to_datetime(df['Date']).dt.day                         # Creating new cloumn of day and month
df['Month'] = pd.to_datetime(df['Date']).dt.month
df.head(11)
```

```python
df.groupby(['Day']).sum()['Total'].plot()
plt.grid(axis='x')
```

#  7. Which month has most sales?

```python
df.groupby(['Month']).sum()['Total'].plot(kind='bar')
plt.grid(axis='y')
```

# 8. Which hours has more sales?

```python
df['Time']
```

```python
pd.to_datetime(df['Time'])  # Converting data type object to datetime format
```

```python
df['Hours'] = pd.to_datetime(df['Time']).dt.hour
df.head(11)
```

```python
df.groupby(['Hours']).sum()['Total'].plot(kind='bar')
plt.grid(axis='y')
```

# 9. Which time people make more e-wallet and cash payment?

```python
df.groupby(['Payment','Hours']).count()['Customer type'].unstack(level=0).plot(kind='bar')
plt.grid(axis='y')
plt.legend(loc='lower right')
```

# 10. Which city has the best rating?

```python
rating = df.groupby(['City']).mean()['Rating']
rating.plot(kind='bar')
plt.grid(axis='y')
```
